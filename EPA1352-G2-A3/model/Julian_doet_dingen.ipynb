```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
G = nx.MultiGraph()
G.add_nodes_from([1,2,3,4,5])
G.add_edge(1,2, '12a')
G.add_edge(1,2, '12b')
G.add_edge(2,3)
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
G = nx.Graph()
G.add_node(1, pos=(12,4))
G.add_node(2, pos=(5,5))
G.add_node(3, pos=(8,7))
G.add_node(4, pos=(3,9))
G.add_node(5, pos=(4,1))
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(1,4)
G.add_edge(2,3)
G.add_edge(2,4)
G.add_edge(2,5)
pos = nx.get_node_attributes(G, 'pos')
nx.draw(G,pos, with_labels=True)
plt.show()
```

```python
cen = nx.betweenness_centrality(G)
cen
```

```python
for node, value in cen.items():
    print('Node:', node, value)
```
